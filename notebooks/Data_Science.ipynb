# Adding a new column using Assign()
## Use when
* Add multiple columns in a single line of code
* When you need to overwrite the values of existing columns [BEST PRACTICE]

```python
score1 = np.random.randint(1,100, size = 1000)
score2 = np.random.randint(1,100, size = 1000)
score3 = np.random.randint(100,201, size = 1000)
```

```python
series1 = pd.Series(score1, index = np.arange(0,1000)) # A column made from score1 values added to a column name called series1
series2 = pd.Series(score2, index = np.arange(0,1000))
series3 = pd.Series(score3, index = np.arange(0,1000))
```

```python
df_exams = df_exams.assign(score1 = series1, score2 = series2)
del df_exams['score3'] 
df_exams.insert(2,'score3', series3)

# adding 2 columns called 'score1' and 'score2'
# Need to reassign the DataFrame
```

```python
df_exams
```
