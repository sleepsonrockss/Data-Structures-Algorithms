# Creating a conditional column for more than 2 choices using np.select()

```python
# create an array based on multiple price tiers (2+)
# create conditions and values variables

conditions = [
    df_laptops['Price_euros'] > 3000, 
    (df_laptops['Price_euros'] > 2000) & (df_laptops['Price_euros'] <= 3000),
    (df_laptops['Price_euros'] > 800) & (df_laptops['Price_euros'] <= 2000),
    df_laptops['Price_euros'] <= 800
]
values = ['Too Expensive', 'Expensive', 'Affordable', 'Cheap']
```

```python
df_laptops['Price_tiers'] = np.select(conditions, values, default = '')
```

```python
df_laptops[df_laptops['Price_tiers'] == 'Too Expensive'].head()
```

# Drop Duplicates using .drop_duplicates()

```python
# drop duplicates in 2 or more columns
df_laptops.drop_duplicates(['Company'])[['Company', 'Price_euros']]
```

```python
# find cheapest and most expensive laptops
df_laptops = df_laptops.sort_values(['Company', 'Price_euros'])
```

```python
df_laptops.drop_duplicates(['Company'], keep = 'first')[['Company', 'Price_euros']] #cheapest laptops for each company
```

```python
#most expensive laptops
df_laptops.drop_duplicates(['Company'], keep = 'last', ignore_index = True)[['Company', 'Price_euros']]
```
