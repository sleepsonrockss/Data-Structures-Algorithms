## re.search()

```python
text = """Hello World.
I love Python!!!!!
My first lucky number is 987-654-321
My second lucky number is 876-543-210
My third lucky number is 765-432-100
My fourth lucky number is 123-456-123-123-456
"""
```

```python
# return the first match digit, match = '9'
re.search(r'\d', text)
```

```python
#flags
re.search(r'^hello', text, flags = re.I) # 'I'(Insensitive) won't consider the query's case
```

## re.findall()

```python
# return all matches
re.findall(r'\d', text) #all digits in the text variable
```

```python
# Exercise
# Match only punctuation
text = """Hello World.
I love Python!!!!!
My first lucky number is 987-654-321
My second lucky number is 876-543-210
My third lucky number is 765-432-100
My fourth lucky number is 123-456-123-123-456
"""
re.findall(r'[^\w\s]',text)
```

```python
# Match the right date format (MM/DD/YYYY

text = '''
13-04-2021
2021-13-04
2021-04-13
'''
re.findall(r'\d{2}-\d{2}-\d{4}',text)
```

```python
# Match the right username format (4-14 characters and only letters or digits)
text = '''
username10
abc
10'''
re.findall(r'\w{4,14}',text)
```
