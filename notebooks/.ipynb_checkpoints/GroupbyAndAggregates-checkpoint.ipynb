```python
import pandas as pd
```

```python
df_cars = pd.read_csv("Car_sales.csv")
```

```python
df_cars = df_cars[['Manufacturer', 'Sales_in_thousands', 'Vehicle_type','Price_in_thousands','Engine_size','Horsepower','Fuel_capacity']]
df_cars
```

```python
# calculate the number of unique vaues
df_cars.nunique()
```

# Agg() method

```python
df_cars.agg('sum')
```

```python
#calculating the mean of numeric data
df_cars.dtypes
df_cars_float = df_cars.select_dtypes(include = ['float64']).agg(['mean'])
```

```python
df_cars.select_dtypes(include = ['float64']).agg(['mean', 'count'])
```

```python
# Calculate the different aggregation per column: sum and mean of 'Sales_in_thousands' and Price_in_thousands sum and max

df_cars_float.agg({'Sales_in_thousands': ['sum', 'mean'], 'Price_in_thousands': ['sum', 'max'] })
```

```python
df_cars[['Sales_in_thousands', 'Price_in_thousands']].agg(['sum'], axis = 1)
```

```python
df_cars['Sales_in_thousands'] + df_cars['Price_in_thousands'] # no. 2 is Nan
```

```python
# Aggregate different function and rename the index of the resulting DataFrame
df_cars.agg(x = ('Sales_in_thousands', 'sum'),
           y = ('Price_in_thousands', 'sum'))
```

# Split - Apply - Combine Strategy

```python
# Split data into seperate groups
passenger_filter = df_cars['Vehicle_type'] == 'Passenger'
car_filter = df_cars['Vehicle_type'] == 'Car'
# df_cars['Vehicle_type'].value_counts()
```

```python
passenger_filter
```

```python
# Apply an operation
passenger_average = df_cars[passenger_filter]['Sales_in_thousands'].mean()
cars_average = df_cars[car_filter]['Sales_in_thousands'].mean()
cars_average # average price of a car vehicle type
```

```python
pd.DataFrame({'Vehicle_type':['Car', 'Passenger'], 'Sales in thousands':[cars_average, passenger_average]}).set_index('Vehicle_type')
```

# GroupBy() Method

```python
# group by Vehicle_type and find the mean
df_cars.groupby(['Vehicle_type'], as_index = False)[[ 'Sales_in_thousands', 'Price_in_thousands', 'Engine_size', 'Horsepower', 'Fuel_capacity']].mean()
```

```python
# group by Manufacturer
groupby_obj = df_cars.groupby(['Manufacturer'], as_index = False)
groupby_obj.groups # Acura has index 0,1,2,3
```

```python
groupby_obj.get_group(('Ford'))
```

```python
df_cars.groupby('Manufacturer', as_index = False)[[ 'Sales_in_thousands', 'Price_in_thousands', 'Engine_size', 'Horsepower', 'Fuel_capacity']].mean()
```

```python
df_cars.groupby('Manufacturer', as_index = False)[[ 'Sales_in_thousands', 'Price_in_thousands', 'Engine_size', 'Horsepower', 'Fuel_capacity']].sum()
```

```python
# find all Null values in columns
df_cars.isnull().sum()
```

```python
# groupby Engine-Size and count values without dropping NaN values
df_cars.groupby('Engine_size', dropna = False)[[ 'Sales_in_thousands', 'Price_in_thousands', 'Engine_size', 'Horsepower', 'Fuel_capacity']].count()
```

# Groupby() and Agg()

```python
# find min and max in each column
df_cars.groupby('Vehicle_type')[[ 'Sales_in_thousands', 'Price_in_thousands', 'Engine_size', 'Horsepower', 'Fuel_capacity']].min()
df_cars.groupby('Vehicle_type')[[ 'Sales_in_thousands', 'Price_in_thousands', 'Engine_size', 'Horsepower', 'Fuel_capacity']].max()
```

```python
df_cars.groupby('Vehicle_type')[[ 'Sales_in_thousands', 'Price_in_thousands', 'Engine_size', 'Horsepower', 'Fuel_capacity']].agg(['min', 'max'])
```

```python
# Set a name for an aggregated column
# min of engine size and max of horsepower
df_cars.groupby('Vehicle_type', as_index = False)[[ 'Engine_size', 'Horsepower']].agg(min_engine_size = ('Engine_size','min'),max_horsepower = ('Horsepower','max'))
```

```python
# calculate the sum of Sales_in_thousands and mean of Price_in_thousands
df_cars.groupby('Manufacturer', as_index = False).agg(sum_Sales = ('Sales_in_thousands','sum'),mean_Price = ('Price_in_thousands','mean'))
```

```python
df_cars.groupby('Manufacturer', as_index = False)['Sales_in_thousands'].sum()
df_cars.groupby('Manufacturer', as_index = False)['Price_in_thousands'].mean()
```

# Lambda

```python
df_cars.groupby('Manufacturer')[[ 'Sales_in_thousands', 'Price_in_thousands', 'Engine_size', 'Horsepower', 'Fuel_capacity']].sum().head()
```

```python
df_cars.groupby('Manufacturer')[[ 'Sales_in_thousands', 'Price_in_thousands' ]].sum().apply(lambda x : x * 1000).rename(columns= {'Sales_in_thousands':'Sales','Price_in_thousands': 'Price'})
df_cars['Sales_in_thousands'] = df_cars['Sales_in_thousands'].apply(lambda x:x*1000)
df_cars.rename(columns = {'Sales_in_thousands':'Sales'})
```
