```python
import pandas as pd
```

```python
df_movies = pd.read_csv('IMDb movies.csv', low_memory=False)
df_movies = df_movies[['imdb_title_id', 'title', 'year', 'genre', 'country']]
df_movies
```

```python
df_ratings = pd.read_csv('IMDb ratings.csv', low_memory=False)
df_ratings = df_ratings[['imdb_title_id','total_votes', 'mean_vote']]
df_ratings
```

# Concat() - Vertically(stacked, need common columns) and Horizontally

![Alt text](./Concat.png)

### Concatenate Vertically

```python
df1 = pd.DataFrame({'id': ['A', 'B', 'C', 'D'],
                   'age': [30, 90, 78, 80]})
df2 = pd.DataFrame({'id': ['E', 'F', 'G', 'H'],
                   'age': [30, 90, 78, 80]})
df = pd.concat([df1, df2], axis = 0, ignore_index = True)
df
```

```python
# extract 50% of the sample from df_movies
df_sample = df_movies.sample(frac=0.5, random_state = 1)
df_sample
```

```python
df_sample.shape
```

```python
df_concat_vertically = pd.concat([df_movies, df_sample], axis = 0, ignore_index = True)
df_concat_vertically
```

```python
df_concat_vertically.shape
```

# Concatenate Horizontally (next to each other) - concats based on a common column

```python
df1 = pd.DataFrame({'id': ['A', 'B', 'C', 'D'],
                   'age': [30, 90, 78, 80]})
df2 = pd.DataFrame({'Job': ['Doctor', 'Statistician', 'Accountant', 'Developer']})
df = pd.concat([df1, df2], axis = 1)
df
```

```python
# df_movies.set_index('imdb_title_id', inplace = True)
# df_ratings.set_index('imdb_title_id', inplace = True)
df_concat_horizontally = pd.concat([df_movies, df_ratings], axis = 1, sort = True)
df_concat_horizontally.head()
```

```python
df_concat_horizontally.shape
```

# Joins using merge()

### Inner Join - Intersection between dataframes (Common Columns)

```python
df1 = pd.DataFrame({'id': ['A', 'B', 'C', 'D'],
                   'age': [30, 90, 78, 80]})
df2 = pd.DataFrame({'id': ['C', 'D', 'G', 'H'],
                   'Job': ['Doctor', 'Statistician', 'Accountant', 'Developer']})
df_inner_join = df1.merge(df2,on = 'id', how = 'inner' )
df_inner_join
```

```python
# Exercise
df_merge = df_movies.merge(df_ratings, on = 'imdb_title_id', how = 'inner')
pd.merge(df_movies, df_ratings, on= 'imdb_title_id')
```

# Full Join (Outer) - Union

```python
df1.merge(df2, on = 'id', how = 'outer')
```

```python
df1.merge(df2, on = 'id', how = 'outer').shape
```

```python
# Exercise
df_movies.merge(df_ratings, on = 'imdb_title_id', how = 'outer')
```

# Exclusive Full Join
![Alt text](./Exclusive_Full_Join.png)

```python
df1.merge(df2, on = 'id', how = 'outer')
```

```python
df1.merge(df2, on = 'id', how = 'outer', indicator = True) # makes a column called _merge which shows which row are in left, right, both
```

```python
df1.merge(df2, on = 'id', how = 'outer', indicator = True).query("_merge == 'left_only' or _merge == 'right_only' ")
```

```python
#exercise
df_movies.merge(df_ratings, on = 'imdb_title_id', how = 'outer', indicator = True).query("_merge == 'left_only' or _merge == 'right_only' ")
```

# Left Join 
![Alt text](./Left_Join.png)

```python
df1 = pd.DataFrame({'id': ['A', 'B', 'C', 'D'],
                   'age': [30, 90, 78, 80]})
df2 = pd.DataFrame({'id': ['C', 'D', 'E', 'F'], 'Job': ['Doctor', 'Statistician', 'Accountant', 'Developer']})
df1.merge(df2, on = 'id', how = 'left')
```

```python
# Exercise
df_movies_sample = df_movies.sample(frac = 0.5, random_state = 1)
df_left = df_movies_sample.merge(df_ratings, on = 'imdb_title_id', how = 'left')
```

```python
df_left.shape
```

# Exclusive Left Join
![Alt text](./Exclusive_Left_Join.png)

```python
df1 = pd.DataFrame({'id': ['A', 'B', 'C', 'D'],
                   'age': [30, 90, 78, 80]})
df2 = pd.DataFrame({'id': ['C', 'D', 'E', 'F'], 'Job': ['Doctor', 'Statistician', 'Accountant', 'Developer']})
df1.merge(df2, on = 'id', how = 'outer', indicator = True).query("_merge == 'left_only'")
```
