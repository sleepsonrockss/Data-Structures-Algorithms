```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
df_population_raw = pd.read_csv('population_total.csv')
```

```python
df_population_raw
```

# Making a Pivot table

```python
# dropping null values
df_population_raw.dropna(inplace = True)
```

```python
# making a pivot table
df_pivot = df_population_raw.pivot(index = 'year', columns = 'country', values = 'population')
df_pivot
```

```python
df_pivot = df_pivot[['United States', 'India', 'China', 'Indonesia', 'Brazil']]
```

```python
df_pivot
```

# Line Plot

```python
df_pivot.plot(kind = 'line', xlabel = 'Year', ylabel = 'Population', title = 'Population from (1955-2020)', figsize = (8,4))
```

# Bar Plot

```python
# select only one year
df_pivot_2020 = df_pivot[df_pivot.index.isin([2020])]
df_pivot_2020
```

```python
# transpose dataframe from rows to columns or vice versa
df_pivot_2020 = df_pivot_2020.T
```

```python
df_pivot_2020.plot(kind = 'bar', color = 'Red', xlabel = 'Country', ylabel = 'Population', title = 'Population in 2020')
```

# Barplot grouped by n variable

```python
df_pivot_sample = df_pivot.loc[[1980, 1990, 2000, 2010, 2020]]
df_pivot_sample
```

```python
# making a grouped barplot
df_pivot_sample.plot(kind = 'bar', xlabel = 'Year', ylabel = 'Population')
```

# Pie Charts

```python
#changing clumn name
df_pivot_2020.rename(columns={2020:'2020'}, inplace = True) # chnage interger column to a string
df_pivot_2020.plot(kind = 'pie', y = '2020', title = 'Population in 2020', table = True) #y is the column that has the data.
```
