# Creating Random Samples

```python
df = pd.read_csv('players_20.csv')
df.set_index('short_name', inplace = True)
df = df[['long_name','age','dob','height_cm','weight_kg','nationality','club']]
df
```

# sample()

```python
# extract 10 random samples from the nationality column
df['nationality'].sample(10, random_state = 99)
```

```python
# extract a random 20% sample from the dataframe
df.sample(frac = 0.2, random_state = 12)
```

```python
# upsample: increasing the sampling rate -> frac > 1 (NOTE : replace parameter should be set to TRUE for frac > 1.)
df.sample(frac = 2, replace = True, random_state = 9) # doubling the number of rows
```

# Filter a dataframe using query()

```python
df.query("age > 34")
#equivalent boolean slicing
df[df['age'] > 34]
```

```python
df.query('age > 34 & nationality == "Italy"')
df[(df['age'] > 34) & (df['nationality'] == 'Italy')] #equivalent boolean slicing
```

```python
df.query("not(age > 34)") #age less then 34
df[~(df['age'] > 34)] #equivalent boolean slicing
```

```python
# convert height to meters and select those with height aboce 1.8
df.query("height_cm/100 > 1.8")
df[df['height_cm']/100 > 1.8]
```

```python
# select players that were born after 1990

# check out data types
df.dtypes
# convert 'dob' column to datetime type
df['dob'] = df['dob'].astype('datetime64[ns]')
df['dob'].dt.year
```

```python
df.query('dob.dt.year > 1990')
df[df['dob'].dt.year > 1990] # boolean equivalent slicing
```

# Making a copy of a dataframe using copy()

### deep = True

```python
df_copy = df.copy() #Deep = True by default (Modifications in the dataframe is not going to be reflected in the copied dataframe.)
```

```python
df.loc['L. Messi', 'height_cm'] = 180
```

```python
df_copy # doesn't change Messi's height
```

### deep = False

```python
df_shallow_copy = df.copy(deep = False)
```

```python
df.loc['Cristiano Ronaldo', 'height_cm'] = 200
```

```python
df_shallow_copy # C Ronaldo height is changed in shallow copy when original is changed
```
